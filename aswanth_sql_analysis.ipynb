```python
# Complete SQL Analysis - All 6 Queries with Output
import pandas as pd
from sqlalchemy import create_engine

# Connect to MySQL
engine = create_engine('mysql+pymysql://root:Aswanth%40880101@localhost/retail_analysis')
print("✓ Connected to MySQL!")

# ================================================
# QUERY 1: TOTAL REVENUE ANALYSIS
# ================================================
print("\n=== QUERY 1: TOTAL REVENUE ANALYSIS ===")
q1 = pd.read_sql("""
SELECT 
    SUM(Weekly_Sales) AS Total_Revenue,
    AVG(Weekly_Sales) AS Avg_Weekly_Sales,
    MAX(Weekly_Sales) AS Best_Week_Sales,
    COUNT(*) AS Total_Records
FROM sales
""", engine)
print(q1.to_string())

# ================================================
# QUERY 2: TOP 10 STORES BY REVENUE
# ================================================
print("\n=== QUERY 2: TOP 10 STORES BY REVENUE ===")
q2 = pd.read_sql("""
SELECT 
    s.Store,
    st.Type AS Store_Type,
    st.Size AS Store_Size,
    SUM(s.Weekly_Sales) AS Total_Revenue,
    AVG(s.Weekly_Sales) AS Avg_Weekly_Sales
FROM sales s 
JOIN stores st ON s.Store = st.Store
GROUP BY s.Store, st.Type, st.Size
ORDER BY Total_Revenue DESC
LIMIT 10
""", engine)
print(q2.to_string())

# ================================================
# QUERY 3: TOP 10 DEPARTMENTS BY REVENUE
# ================================================
print("\n=== QUERY 3: TOP 10 DEPARTMENTS BY REVENUE ===")
q3 = pd.read_sql("""
SELECT 
    Dept,
    SUM(Weekly_Sales) AS Total_Revenue,
    AVG(Weekly_Sales) AS Avg_Weekly_Sales,
    COUNT(*) AS Total_Records
FROM sales 
GROUP BY Dept
ORDER BY Total_Revenue DESC
LIMIT 10
""", engine)
print(q3.to_string())

# ================================================
# QUERY 4: HOLIDAY VS NON-HOLIDAY SALES
# ================================================
print("\n=== QUERY 4: HOLIDAY VS NON-HOLIDAY SALES ===")
q4 = pd.read_sql("""
SELECT 
    CASE 
        WHEN IsHoliday = 1 THEN 'Holiday'
        ELSE 'Non-Holiday'
    END AS Holiday_Type,
    SUM(Weekly_Sales) AS Total_Revenue,
    AVG(Weekly_Sales) AS Avg_Weekly_Sales,
    COUNT(*) AS Total_Weeks
FROM sales 
GROUP BY IsHoliday
ORDER BY Avg_Weekly_Sales DESC
""", engine)
print(q4.to_string())

# ================================================
# QUERY 5: REVENUE BY STORE TYPE
# ================================================
print("\n=== QUERY 5: REVENUE BY STORE TYPE ===")
q5 = pd.read_sql("""
SELECT 
    st.Type AS Store_Type,
    COUNT(DISTINCT s.Store) AS Number_of_Stores,
    SUM(s.Weekly_Sales) AS Total_Revenue,
    AVG(s.Weekly_Sales) AS Avg_Weekly_Sales
FROM sales s 
JOIN stores st ON s.Store = st.Store
GROUP BY st.Type
ORDER BY Total_Revenue DESC
""", engine)
print(q5.to_string())

# ================================================
# QUERY 6: MONTHLY SALES TREND
# ================================================
print("\n=== QUERY 6: MONTHLY SALES TREND ===")
q6 = pd.read_sql("""
SELECT 
    YEAR(Date) AS Year,
    MONTH(Date) AS Month,
    SUM(Weekly_Sales) AS Monthly_Revenue,
    AVG(Weekly_Sales) AS Avg_Weekly_Sales
FROM sales 
GROUP BY YEAR(Date), MONTH(Date)
ORDER BY Year, Month
""", engine)
print(q6.to_string())

print("\n✓ All 6 SQL queries executed successfully!")
print("✓ Analysis complete!")
```
